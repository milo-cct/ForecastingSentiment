```python
import nltk

import warnings
warnings.filterwarnings('ignore')

from nltk.corpus import stopwords
nltk.download('stopwords')

import string
from nltk.stem import PorterStemmer
```

### Language processing/cleaning

```python
# Store the stopwords into the object named as "stop_words"
stop_words = stopwords.words('english')

# Store the string.punctuation into an object punct
punct = string.punctuation

# Initialise an object using a method PorterStemmer
stemmer = PorterStemmer()
```

```python
X = df['text']
```

```python
import re

cleaned_data=[]

# For loop from first value to length(X), ^a-zA-Z means include small and capital case letters

for i in range(len(X)):
    tweet = re.sub('[^a-zA-Z]', ' ', X.iloc[i])
    tweet = tweet.lower().split()
    tweet = [stemmer.stem(word) for word in tweet if (word not in stop_words) and (word not in punct)]
    tweet = ' '.join(tweet)
    cleaned_data.append(tweet)
```

```python
df['clean']=pd.DataFrame(cleaned_data)
```

following method adapted from https://medium.com/@aaabulkhair/data-imputation-demystified-time-series-data-69bc9c798cb7

```python
# Make a copy of the original DataFrame
df_copy = pos_domain.copy()

# Mark the missing values before imputation
imputed_indices = df_copy[df_copy['mean'].isnull()].index


# Apply the rolling mean imputation method
df_copy['mean'] = df_copy['mean'].fillna(df_copy['mean'].rolling(window=5, min_periods=1).mean().shift(1))

# Plot the main line with markers
df_copy['mean'].plot(style='.-', figsize=(12,8), title='Average tweet sentiment')

# Add points where data was imputed with red color
plt.scatter(imputed_indices, df_copy.loc[imputed_indices, 'mean'], color='red')

# Set labels
plt.xlabel('Time')
plt.ylabel('Mean daily sentiment')

plt.show()
```

playing around with window values the result of this method is not awful. in fact all of these look good aside from the ones in the very fist gap, which should be more influenced by the values to the right. let's figure out how to apply the 5% reduction and use this method provisionally before continuing. interestingly, using this method twice, on top of itself, may prove to be especially effective. The hyperparameters were adjusted, finding that a minimum combination of 10 between the windows was needed for an coverage, and finding that negative shifts added the possibility for some level of backwards filling which was ideal for the first gap

```python
df_copy = pos_domain.copy()

# Mark the missing values before imputation
imputed_indices = df_copy[df_copy['mean'].isnull()].index


# Apply the rolling mean imputation method
df_copy['mean'] = df_copy['mean'].fillna(df_copy['mean'].rolling(window=6, min_periods=1).mean().shift(1))
df_copy['mean'] = df_copy['mean'].fillna(df_copy['mean'].rolling(window=4, min_periods=1).mean().shift(0))

df_copy.loc[imputed_indices, 'mean'] = df_copy.loc[imputed_indices, 'mean'] * 0.95

# Plot the main line with markers
df_copy['mean'].plot(style='.-', figsize=(12,8), title='Average tweet sentiment')

# Add points where data was imputed with red color
plt.scatter(imputed_indices, df_copy.loc[imputed_indices, 'mean'], color='red')

# Set labels
plt.xlabel('Time')
plt.ylabel('Mean daily sentiment')

plt.show()
```

```python
df_copy = pos_domain.copy()

# Mark the missing values before imputation
imputed_indices = df_copy[df_copy['mean'].isnull()].index


# Apply the rolling mean imputation method
df_copy['mean'] = df_copy['mean'].fillna(df_copy['mean'].rolling(window=4, min_periods=1).mean().shift(1))
df_copy['mean'] = df_copy['mean'].fillna(df_copy['mean'].rolling(window=7, min_periods=1).mean().shift(-2))

df_copy.loc[imputed_indices, 'mean'] = df_copy.loc[imputed_indices, 'mean'] * 0.95

# Plot the main line with markers
df_copy['mean'].plot(style='.-', figsize=(12,8), title='Average tweet sentiment')

# Add points where data was imputed with red color
plt.scatter(imputed_indices, df_copy.loc[imputed_indices, 'mean'], color='red')

# Set labels
plt.xlabel('Time')
plt.ylabel('Mean daily sentiment')

plt.show()
```

### This is the pattern that would be used for prediction in the case of inversion permanency

```python
# SARIMA example
from statsmodels.tsa.statespace.sarimax import SARIMAX

# fit model
model = SARIMAX(nodomain['negmeans'], order=(1, 1, 1), seasonal_order=(0, 0, 0, 0))
model_fit = model.fit(disp=False)
# make prediction
yhat = model_fit.predict(len(nodomain), len(nodomain))
print(yhat) 
```

```python
model = SARIMAX(nodomain['negmeans'], order=(2, 1, 2), seasonal_order=(0, 2, 0,30))
model_fit = model.fit(disp=False)
# make prediction

forecast = model_fit.get_forecast(steps=30)
predicted_values = forecast.predicted_mean

predicted_values.plot(kind='line')
```
